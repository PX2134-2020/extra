```
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```
#F coil
Fy = np.array([30,26.1,20.6,40.2,50,56.7,73.3,75.8])
Fx = np.array([0.55,0.465,0.379,0.724,0.907,1.021,1.312,1.36])

Fy = Fy*(10**6)
Fx = Fx/2

plt.plot(Fx,Fy,'.',label='Data Points')

F_coeff = polyfit( Fx, Fy, 1 ) 
p = poly1d( F_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( Fx, p(Fx), label='Best Fit Line' ) 

plt.grid()
plt.title('Resonance dip of F coil')
plt.xlabel('Current / A')
plt.ylabel('frequency / Hz')
plt.legend()

print('line of best fit, y =',F_coeff[0],'x +',F_coeff[1],'\n so gradient =',F_coeff[0],'Hz/A')

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

Fgs = (F_coeff[0]*2*np.pi*r*h*5*(5**0.5))/(muB*mu0*n*8)

print('\nthe g-factor, gs =',Fgs)
```

```
#E coil
Ey = np.array([10.5,28.9,35.4,24.3,20.1,15])
Ex = np.array([0.194,0.533,0.642,0.441,0.37,0.276])

Ey = Ey*(10**6)
Ex = Ex/2

plt.plot(Ex,Ey,'.',label='Data Points')

E_coeff = polyfit( Ex, Ey, 1 ) 
p = poly1d( E_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( Ex, p(Ex), label='Best Fit Line' ) 

plt.grid()
plt.title('Resonance dip of E coil')
plt.xlabel('Current / A')
plt.ylabel('frequency / Hz')
plt.legend()

print('line of best fit, y =',E_coeff[0],'x +',E_coeff[1],'\n so gradient =',E_coeff[0],'Hz/A')

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

Egs = (E_coeff[0]*2*np.pi*r*h*5*(5**0.5))/(muB*mu0*n*8)

print('\nthe g-factor, gs =',Egs)
```

```
#G coil
Gy = np.array([51.5,129.2,115,100.7,88.9,79.9,60.4,69.1])
Gx = np.array([0.922,2.3,2.05,1.768,1.571,1.41,1.072,1.23])

Gy = Gy*(10**6)
Gx = Gx/2

plt.plot(Gx,Gy,'.',label='Data Points')

G_coeff = polyfit( Gx, Gy, 1 ) 
p = poly1d( G_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( Gx, p(Gx), label='Best Fit Line' ) 

plt.grid()
plt.title('Resonance dip of G coil')
plt.xlabel('Current / A')
plt.ylabel('frequency / Hz')
plt.legend()

print('line of best fit, y =',G_coeff[0],'x +',G_coeff[1],'\n so gradient =',G_coeff[0],'Hz/A')

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

Ggs = (G_coeff[0]*2*np.pi*r*h*5*(5**0.5))/(muB*mu0*n*8)

print('\nthe g-factor, gs =',Ggs)
```

```
g = 2.0023193043622 

print('Error G_gs =', (g-Ggs)/g)
print('Error F_gs =', (g-Fgs)/g)
print('Error E_gs =', (g-Egs)/g)
print('Error gs   =',g*0.15)
print('lower bound of expected gs =',g-g*0.15)
```

```
#G coil
f_y = np.array([30,50,75,60,40,21])
H_x = np.array([2.5,3.75,4.4,3.7,3,1.85])
W_x = np.array([0.9,0.95,0.9,0,0.85,0])

E = 2*np.pi*h*f_y*(10**6)

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

plt.figure(0)
plt.plot(H_x,E,'.',label='Energy vs Height of peak')

G_coeff = polyfit( H_x,E, 1 ) 
p = poly1d( G_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( H_x, p(H_x), label='Best Fit Line' ) 

plt.grid()
plt.title('F coil Energy relation investigation')
plt.xlabel('Height of peak / V')
plt.ylabel('Energy / J')
plt.legend()

print('line of best fit, E =',G_coeff[0],'Height +',G_coeff[1],'\n so gradient =',G_coeff[0],'Hz/A')

plt.figure(1)
plt.plot(W_x,E,'.',label='Energy vs FWHM')

E_coeff = polyfit( W_x,E, 1 ) 
E = poly1d( E_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( W_x, E(W_x), label='Best Fit Line' ) 

plt.grid()
plt.title('F coil Energy relation investigation')
plt.xlabel('FWHM / V')
plt.ylabel('Energy / J')
plt.legend()

print('line of best fit, E =',E_coeff[0],'FWHM +',E_coeff[1],'\n so gradient =',E_coeff[0],'Hz/A')

```
