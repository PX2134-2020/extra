```
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```
#F coil
Fy = np.array([30,26.1,20.6,40.2,50,56.7,73.3,75.8])
Fx = np.array([0.55,0.465,0.379,0.724,0.907,1.021,1.312,1.36])

Fy = Fy*(10**6)
Fx = Fx/2

plt.plot(Fx,Fy,'.',label='Data Points')

F_coeff = polyfit( Fx, Fy, 1 ) 
p = poly1d( F_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( Fx, p(Fx), label='Best Fit Line' ) 

plt.grid()
plt.title('Resonance dip of F coil')
plt.xlabel('Current / A')
plt.ylabel('frequency / Hz')
plt.legend()

print('line of best fit, y =',F_coeff[0],'x +',F_coeff[1],'\n so gradient =',F_coeff[0],'Hz/A')

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

Fgs = (F_coeff[0]*2*np.pi*r*h*5*(5**0.5))/(muB*mu0*n*8)

print('\nthe g-factor, gs =',Fgs)
```

```
#E coil
Ey = np.array([10.5,28.9,35.4,24.3,20.1,15])
Ex = np.array([0.194,0.533,0.642,0.441,0.37,0.276])

Ey = Ey*(10**6)
Ex = Ex/2

plt.plot(Ex,Ey,'.',label='Data Points')

E_coeff = polyfit( Ex, Ey, 1 ) 
p = poly1d( E_coeff ) 
x = linspace( 0, 1, 1000 ) 
plot( Ex, p(Ex), label='Best Fit Line' ) 

plt.grid()
plt.title('Resonance dip of E coil')
plt.xlabel('Current / A')
plt.ylabel('frequency / Hz')
plt.legend()

print('line of best fit, y =',E_coeff[0],'x +',E_coeff[1],'\n so gradient =',E_coeff[0],'Hz/A')

n = 320
mu0 = (4*np.pi)*10**(-7)
muB = 9.274*10**(-24)
r = 6.8*10**(-2)
h = 1.055*10**(-34)

Egs = (E_coeff[0]*2*np.pi*r*h*5*(5**0.5))/(muB*mu0*n*8)

print('\nthe g-factor, gs =',Egs)
```
